```python
# LISTAS — equivalente aproximado a ArrayList<Object> en Java
# Pero sin declarar tipo, sin new, sin <>

# Lista de analitos en un hemograma
analitos_hemograma = ['eritrocitos', 'hemoglobina', 'hematocrito', 
                       'leucocitos', 'plaquetas']

# Acceso por índice — igual que en Java, empieza en 0
print(analitos_hemograma[0])    # 'eritrocitos'
print(analitos_hemograma[-1])   # 'plaquetas' — índice negativo: desde el final
print(analitos_hemograma[1:3])  # ['hemoglobina', 'hematocrito'] — slicing

# len() en lugar de .size() o .length
print(f"Número de analitos: {len(analitos_hemograma)}")

# Agregar elementos — .append() existe en ambos lenguajes
analitos_hemograma.append('VCM')
print(analitos_hemograma)

# Verificar si un elemento está en la lista — Python tiene operador 'in'
# En Java harías lista.contains(elemento)
if 'hemoglobina' in analitos_hemograma:
    print("La hemoglobina está en el panel")
```

```python
# Las listas pueden contener elementos de distintos tipos
# En Java esto requería List<Object> o genéricos complejos

resultado_muestra = ['C001', 'Labrador', 3, 15.1, True, None]
#                    id      raza      edad  Hgb   normal  comentario

# Pero en análisis de datos, las listas homogéneas son más útiles
# Las listas mixtas las verás más en scripts generales, no en análisis

# Iterar sobre una lista — sintaxis más limpia que Java
valores_hematocrito = [45.2, 35.4, 51.8, 28.6, 48.0]

print("Valores de hematocrito:")
for valor in valores_hematocrito:
    print(f"  {valor}%")

# enumerate() da índice + valor — equivale al for con contador de Java
print("\nCon índice:")
for indice, valor in enumerate(valores_hematocrito):
    print(f"  Paciente {indice + 1}: {valor}%")
```

```python
# Diccionario de diccionarios — estructura muy común en datos clínicos
# Equivale a un Map<String, Map<String, Object>> en Java

rangos_referencia_canino = {
    'eritrocitos':  {'min': 5.5,   'max': 8.5,   'unidad': 'millones/µL'},
    'hemoglobina':  {'min': 12.0,  'max': 18.0,  'unidad': 'g/dL'},
    'hematocrito':  {'min': 37.0,  'max': 55.0,  'unidad': '%'},
    'VCM':          {'min': 60.0,  'max': 77.0,  'unidad': 'fL'},
    'HCM':          {'min': 19.5,  'max': 24.5,  'unidad': 'pg'},
    'CHCM':         {'min': 32.0,  'max': 36.0,  'unidad': 'g/dL'},
    'leucocitos':   {'min': 6.0,   'max': 17.0,  'unidad': 'miles/µL'},
    'plaquetas':    {'min': 200.0, 'max': 500.0, 'unidad': 'miles/µL'}
}

# Acceso anidado
hgb_min = rangos_referencia_canino['hemoglobina']['min']
hgb_max = rangos_referencia_canino['hemoglobina']['max']
print(f"Rango de referencia Hgb canino: {hgb_min}–{hgb_max} g/dL")

# Verificar si un analito tiene rango definido
analito_consulta = 'neutrofilos_pct'
if analito_consulta in rangos_referencia_canino:
    print(f"Rango definido para {analito_consulta}")
else:
    print(f"No hay rango definido para {analito_consulta}")
```

```python
# TUPLAS — como listas pero INMUTABLES (no se pueden modificar)
# Úsalas cuando los datos no deben cambiar: coordenadas, constantes, rangos fijos

# Un rango de referencia es un buen candidato para tupla — no debería cambiar
rango_hematocrito_canino = (37.0, 55.0)    # (min, max)
rango_leucocitos_canino  = (6.0,  17.0)

# Desempaquetado — muy útil en Python, no existe en Java
hto_min, hto_max = rango_hematocrito_canino
print(f"Hematocrito de referencia: {hto_min}–{hto_max}%")

# Las tuplas son más eficientes en memoria que las listas
# Para datos que no cambian, prefiere tupla sobre lista

# SETS — colecciones sin duplicados, sin orden
# Equivalen a HashSet en Java — útiles para verificar unicidad

especies_atendidas = {'Canino', 'Felino', 'Canino', 'Bovino', 'Felino', 'Equino'}
print(f"\nEspecies únicas: {especies_atendidas}")  # Los duplicados se eliminan

patogenos_muestra_A = {'E. coli', 'Klebsiella', 'Pseudomonas'}
patogenos_muestra_B = {'Staphylococcus', 'E. coli', 'Proteus'}

# Operaciones de conjuntos — muy útiles en microbiología
print(f"Patógenos comunes: {patogenos_muestra_A & patogenos_muestra_B}")   # intersección
print(f"Todos los patógenos: {patogenos_muestra_A | patogenos_muestra_B}") # unión
print(f"Solo en A: {patogenos_muestra_A - patogenos_muestra_B}")           # diferencia
```

## Funciones — encapsular lógica clínica reutilizable

Una función en Python: `def nombre(parámetros): ... return resultado`  
Sin tipos declarados, sin clase obligatoria, sin llaves — solo indentación.

```python
# Función básica — comparar con Java:
# Java:   public boolean estaEnRango(double valor, double min, double max) { return ... }
# Python: def esta_en_rango(valor, minimo, maximo): return ...

def esta_en_rango(valor, minimo, maximo):
    """
    Verifica si un valor analítico está dentro del rango de referencia.
    
    Parámetros:
        valor   : float — resultado del analito
        minimo  : float — límite inferior del rango de referencia
        maximo  : float — límite superior del rango de referencia
    
    Retorna:
        bool — True si está en rango, False si está fuera
    """
    # Las triples comillas de arriba son docstrings — documentación de la función
    # Equivalen al JavaDoc pero con sintaxis más simple
    return minimo <= valor <= maximo
    # Python permite esta sintaxis de rango — en Java sería: valor >= min && valor <= max


# Probar la función
hgb_paciente = 11.8
en_rango = esta_en_rango(hgb_paciente, 12.0, 18.0)
print(f"Hemoglobina {hgb_paciente} g/dL en rango (12–18): {en_rango}")

hto_paciente = 45.2
print(f"Hematocrito {hto_paciente}% en rango (37–55): {esta_en_rango(hto_paciente, 37.0, 55.0)}")
```

```python
# Parámetros con valor por defecto — no existe en Java estándar
# (Java usa sobrecarga de métodos para esto)

def calcular_cv(valores, decimales=2):
    """
    Calcula el coeficiente de variación (CV%) de una lista de valores.
    Fundamental en control de calidad analítico.
    
    CV% = (desviación estándar / media) × 100
    """
    import statistics     # módulo estándar de Python para estadística básica
    
    media = statistics.mean(valores)
    desv_std = statistics.stdev(valores)
    cv = (desv_std / media) * 100
    
    return round(cv, decimales)   # round() redondea al número de decimales indicado


# Usar con el valor por defecto (2 decimales)
mediciones_glucosa = [5.1, 5.3, 4.9, 5.2, 5.0, 5.1, 4.8, 5.2]
print(f"CV% glucosa: {calcular_cv(mediciones_glucosa)}%")

# Usar con parámetro personalizado
print(f"CV% glucosa (4 decimales): {calcular_cv(mediciones_glucosa, decimales=4)}%")

# Criterio clínico: CV% < 5% es aceptable en hematología automatizada
cv_resultado = calcular_cv(mediciones_glucosa)
print(f"Control de calidad: {'APROBADO ✓' if cv_resultado < 5.0 else 'REVISAR ⚠'}")
```

## Función principal: intérprete automático de hemograma canino

Esta función encapsula la lógica de interpretación clínica que un laboratorista 
aplica al revisar un hemograma. Es reutilizable: puede aplicarse a cualquier 
paciente con los mismos rangos de referencia.

```python
# Primero definimos los rangos de referencia como constante global
# Convención Python: constantes en MAYÚSCULAS (equivale a static final en Java)

RANGOS_CANINO = {
    'eritrocitos':      (5.5,  8.5),    # millones/µL
    'hemoglobina':      (12.0, 18.0),   # g/dL
    'hematocrito':      (37.0, 55.0),   # %
    'VCM':              (60.0, 77.0),   # fL
    'HCM':              (19.5, 24.5),   # pg
    'CHCM':             (32.0, 36.0),   # g/dL
    'leucocitos':       (6.0,  17.0),   # miles/µL
    'neutrofilos_pct':  (60.0, 77.0),   # %
    'linfocitos_pct':   (12.0, 30.0),   # %
    'monocitos_pct':    (3.0,  10.0),   # %
    'eosinofilos_pct':  (0.0,  9.0),    # %
    'plaquetas':        (200.0, 500.0)  # miles/µL
}


def interpretar_analito(nombre, valor, rangos=RANGOS_CANINO):
    """
    Interpreta un analito individual comparando con el rango de referencia.
    
    Retorna un diccionario con: valor, estado y desviación porcentual.
    """
    if nombre not in rangos:
        return {'valor': valor, 'estado': 'Sin referencia', 'desviacion_pct': None}
    
    minimo, maximo = rangos[nombre]   # desempaquetado de tupla
    
    if valor < minimo:
        # Calcular qué tan por debajo está, en porcentaje
        desviacion = ((valor - minimo) / minimo) * 100
        estado = 'BAJO'
    elif valor > maximo:
        desviacion = ((valor - maximo) / maximo) * 100
        estado = 'ALTO'
    else:
        desviacion = 0.0
        estado = 'Normal'
    
    return {
        'valor':          valor,
        'estado':         estado,
        'desviacion_pct': round(desviacion, 1)
    }


# Probar con un analito
resultado = interpretar_analito('hemoglobina', 11.8)
print(f"Hemoglobina 11.8 g/dL → {resultado}")

resultado2 = interpretar_analito('leucocitos', 18.6)
print(f"Leucocitos 18.6 miles/µL → {resultado2}")
```

```python
def interpretar_hemograma_completo(datos_paciente, rangos=RANGOS_CANINO):
    """
    Genera una interpretación clínica completa de un hemograma canino.
    
    Parámetros:
        datos_paciente : dict — resultado completo del paciente
        rangos         : dict — rangos de referencia (por defecto canino adulto)
    
    Retorna:
        dict con resumen clínico, hallazgos por analito y alertas
    """
    
    # Analitos que vamos a evaluar (los que tienen rango definido)
    analitos_a_evaluar = [
        'eritrocitos', 'hemoglobina', 'hematocrito',
        'VCM', 'HCM', 'CHCM',
        'leucocitos', 'neutrofilos_pct', 'linfocitos_pct',
        'monocitos_pct', 'eosinofilos_pct', 'plaquetas'
    ]
    
    hallazgos = {}        # guardará interpretación de cada analito
    alteraciones = []     # lista de analitos fuera de rango
    alertas = []          # alertas clínicas específicas
    
    # Evaluar cada analito
    for analito in analitos_a_evaluar:
        if analito in datos_paciente:
            valor = datos_paciente[analito]
            interpretacion = interpretar_analito(analito, valor, rangos)
            hallazgos[analito] = interpretacion
            
            if interpretacion['estado'] != 'Normal':
                alteraciones.append(analito)
    
    # --- Lógica de alertas clínicas ---
    
    # Alerta: anemia (Hto bajo)
    if 'hematocrito' in hallazgos:
        hto = datos_paciente.get('hematocrito', 0)
        if hto < 25:
            alertas.append('⚠ ANEMIA SEVERA — considerar transfusión')
        elif hto < 37:
            alertas.append('⚠ Anemia')
    
    # Alerta: leucocitosis marcada — posible proceso infeccioso
    if 'leucocitos' in hallazgos:
        leu = datos_paciente.get('leucocitos', 0)
        if leu > 30:
            alertas.append('⚠ LEUCOCITOSIS MARCADA — descartar sepsis')
        elif leu > 17:
            alertas.append('⚠ Leucocitosis — posible proceso infeccioso/inflamatorio')
    
    # Alerta: leucopenia — posible inmunosupresión
        elif leu < 6:
            alertas.append('⚠ Leucopenia — posible inmunosupresión o proceso viral')
    
    # Alerta: trombocitopenia
    if 'plaquetas' in hallazgos:
        plt_val = datos_paciente.get('plaquetas', 999)
        if plt_val < 50:
            alertas.append('⚠ TROMBOCITOPENIA SEVERA — riesgo hemorrágico')
        elif plt_val < 200:
            alertas.append('⚠ Trombocitopenia')
    
    # Alerta: neutrofilia con desviación a la izquierda aproximada
    if 'neutrofilos_pct' in datos_paciente and 'leucocitos' in datos_paciente:
        if datos_paciente['neutrofilos_pct'] > 80 and datos_paciente['leucocitos'] > 15:
            alertas.append('⚠ Neutrofilia marcada — posible infección bacteriana aguda')
    
    # --- Construir diagnóstico orientativo ---
    if not alteraciones:
        diagnostico_orientativo = 'Hemograma dentro de los límites de referencia'
    else:
        diagnostico_orientativo = f"Alteraciones en: {', '.join(alteraciones)}"
    
    # --- Resultado final ---
    return {
        'paciente_id':           datos_paciente.get('paciente_id', 'Desconocido'),
        'especie':               datos_paciente.get('especie', 'Canino'),
        'diagnostico_orientativo': diagnostico_orientativo,
        'alteraciones':          alteraciones,
        'alertas':               alertas,
        'hallazgos':             hallazgos,
        'total_analitos_evaluados': len(hallazgos),
        'total_alteraciones':    len(alteraciones)
    }


print("Función 'interpretar_hemograma_completo' definida ✓")
```

```python
# Probar con un caso de anemia moderada + leucocitosis
paciente_c004 = {
    'paciente_id':      'C004',
    'especie':          'Canino',
    'raza':             'Bulldog',
    'edad_años':        4,
    
    'eritrocitos':      3.9,
    'hemoglobina':      9.5,
    'hematocrito':      28.6,
    'VCM':              73.3,
    'HCM':              24.4,
    'CHCM':             33.2,
    
    'leucocitos':       18.6,
    'neutrofilos_pct':  82,
    'linfocitos_pct':   10,
    'monocitos_pct':    5,
    'eosinofilos_pct':  2,
    
    'plaquetas':        98
}

resultado_c004 = interpretar_hemograma_completo(paciente_c004)

# Mostrar el resultado de forma legible
print(f"{'='*55}")
print(f"INTERPRETACIÓN DE HEMOGRAMA — {resultado_c004['paciente_id']}")
print(f"{'='*55}")
print(f"Especie:  {resultado_c004['especie']}")
print(f"Diagnóstico orientativo: {resultado_c004['diagnostico_orientativo']}")
print(f"\nAnalitos evaluados: {resultado_c004['total_analitos_evaluados']}")
print(f"Con alteración:     {resultado_c004['total_alteraciones']}")

if resultado_c004['alertas']:
    print(f"\nALERTAS CLÍNICAS:")
    for alerta in resultado_c004['alertas']:
        print(f"  {alerta}")

print(f"\nDETALLE POR ANALITO:")
for analito, datos in resultado_c004['hallazgos'].items():
    estado = datos['estado']
    valor  = datos['valor']
    desv   = datos['desviacion_pct']
    
    # Formato condicional según estado
    if estado == 'Normal':
        marcador = '✓'
    elif estado == 'BAJO':
        marcador = f'↓ ({desv}%)'
    else:
        marcador = f'↑ (+{desv}%)'
    
    print(f"  {analito:<20} {valor:>7}   {marcador}")
```

## Comprensión de listas — transformar datos sin bucles verbosos

Es la forma pythónica de crear listas nuevas a partir de listas existentes.
En Java no existe equivalente directo (Streams son lo más cercano, pero más complejos).

```python
# Comparación directa: bucle tradicional vs comprensión de lista

valores_hto = [45.2, 35.4, 51.8, 28.6, 48.0, 37.5, 42.6, 30.8, 49.5, 41.4]

# --- FORMA JAVA / bucle tradicional ---
estados_bucle = []
for valor in valores_hto:
    if valor < 37.0:
        estados_bucle.append('Bajo')
    elif valor > 55.0:
        estados_bucle.append('Alto')
    else:
        estados_bucle.append('Normal')

# --- FORMA PYTHON / comprensión de lista ---
estados_comp = [
    'Bajo'   if valor < 37.0 else
    'Alto'   if valor > 55.0 else
    'Normal'
    for valor in valores_hto
]

# Ambas producen el mismo resultado
print("Con bucle:        ", estados_bucle)
print("Con comprensión:  ", estados_comp)
print(f"Son iguales: {estados_bucle == estados_comp}")
```

```python
# Comprensión con filtro — solo conservar algunos elementos

# Obtener los valores de Hto que están bajo el rango de referencia
htos_bajos = [v for v in valores_hto if v < 37.0]
print(f"Hematocritos bajo referencia (<37%): {htos_bajos}")

# Combinar transformación + filtro
# Obtener los hematocritos bajos y calcular qué tan lejos están del límite inferior
deficit_hto = [round(37.0 - v, 1) for v in valores_hto if v < 37.0]
print(f"Déficit respecto al límite inferior: {deficit_hto} puntos porcentuales")
```

```python
# Aplicación real: procesar una lista de resultados con la función que construimos

# Lista de hemogramas de múltiples pacientes
lista_hemogramas = [
    {'paciente_id': 'C001', 'especie': 'Canino', 'eritrocitos': 6.2,
     'hemoglobina': 15.1, 'hematocrito': 45.2, 'VCM': 72.9, 'HCM': 24.4,
     'CHCM': 33.4, 'leucocitos': 8.5, 'neutrofilos_pct': 68,
     'linfocitos_pct': 22, 'monocitos_pct': 6, 'eosinofilos_pct': 3,
     'plaquetas': 245},
    
    {'paciente_id': 'C002', 'especie': 'Canino', 'eritrocitos': 4.8,
     'hemoglobina': 11.8, 'hematocrito': 35.4, 'VCM': 73.8, 'HCM': 24.6,
     'CHCM': 33.3, 'leucocitos': 12.3, 'neutrofilos_pct': 75,
     'linfocitos_pct': 16, 'monocitos_pct': 5, 'eosinofilos_pct': 3,
     'plaquetas': 189},
    
    {'paciente_id': 'C004', 'especie': 'Canino', 'eritrocitos': 3.9,
     'hemoglobina': 9.5, 'hematocrito': 28.6, 'VCM': 73.3, 'HCM': 24.4,
     'CHCM': 33.2, 'leucocitos': 18.6, 'neutrofilos_pct': 82,
     'linfocitos_pct': 10, 'monocitos_pct': 5, 'eosinofilos_pct': 2,
     'plaquetas': 98}
]

# Procesar todos los hemogramas con comprensión de lista
# Una sola línea reemplaza un bucle con append()
interpretaciones = [interpretar_hemograma_completo(hgb) for hgb in lista_hemogramas]

# Mostrar resumen de todos los pacientes
print(f"{'ID':<8} {'Alteraciones':<15} {'Alertas'}")
print('-' * 50)
for interp in interpretaciones:
    n_alertas = len(interp['alertas'])
    alerta_str = f"{n_alertas} alerta(s)" if n_alertas > 0 else "—"
    print(f"{interp['paciente_id']:<8} {interp['total_alteraciones']:<15} {alerta_str}")
```

```python
# Comprensión de diccionarios — menos frecuente pero muy útil
# Crear un diccionario de {paciente_id: diagnóstico_orientativo}

resumen_diagnosticos = {
    interp['paciente_id']: interp['diagnostico_orientativo']
    for interp in interpretaciones
}

print("Resumen por paciente:")
for paciente, diagnostico in resumen_diagnosticos.items():
    print(f"  {paciente}: {diagnostico}")
```

```python
# Lambda — función de una sola línea sin nombre
# Equivalente a expresiones lambda de Java 8+, pero con sintaxis más simple

# Java:   (double valor) -> valor < 37.0 ? "Bajo" : "Normal"
# Python: lambda valor: 'Bajo' if valor < 37.0 else 'Normal'

clasificar_hto = lambda valor: 'Bajo' if valor < 37.0 else ('Alto' if valor > 55.0 else 'Normal')

# Usar la lambda
print(clasificar_hto(28.6))  # Bajo
print(clasificar_hto(45.2))  # Normal
print(clasificar_hto(58.0))  # Alto

# Usar lambda con map() — aplicar una función a cada elemento de una lista
# map() es equivalente a .stream().map() en Java, pero más simple
valores_hto = [45.2, 35.4, 51.8, 28.6, 48.0]
estados = list(map(clasificar_hto, valores_hto))
print(f"\nEstados: {estados}")

# Nota: en pandas usarás lambda con .apply() — lo veremos en las próximas sesiones
# Es uno de los usos más frecuentes de lambda en análisis de datos real
```

```python
def es_anemia_normocitica_normocromica(datos_paciente):
    """
    Evalúa si un paciente presenta patrón de anemia normocítica normocrómica.
    Típica de enfermedad crónica, inflamación sistémica o insuficiencia renal.
    
    Retorna dict con evaluación de cada criterio y resultado del patrón completo.
    """
    hto  = datos_paciente.get('hematocrito', 999)
    vcm  = datos_paciente.get('VCM', 0)
    chcm = datos_paciente.get('CHCM', 0)
    
    es_anemia       = hto  < 37.0                    # Hto bajo el límite inferior
    es_normocitica  = 60.0 <= vcm  <= 77.0           # VCM en rango
    es_normocromica = 32.0 <= chcm <= 36.0           # CHCM en rango
    patron_completo = es_anemia and es_normocitica and es_normocromica
    
    return {
        'paciente_id':      datos_paciente.get('paciente_id', '?'),
        'es_anemia':        es_anemia,
        'es_normocitica':   es_normocitica,
        'es_normocromica':  es_normocromica,
        'patron_completo':  patron_completo
    }


# Aplicar a todos los pacientes con comprensión de lista
evaluaciones = [es_anemia_normocitica_normocromica(hgb) for hgb in lista_hemogramas]

# Mostrar resultados
print(f"{'ID':<8} {'Anemia':<10} {'Normocít.':<12} {'Normocróm.':<13} {'Patrón completo'}")
print('-' * 58)
for ev in evaluaciones:
    print(f"{ev['paciente_id']:<8} "
          f"{'Sí' if ev['es_anemia']       else 'No':<10} "
          f"{'Sí' if ev['es_normocitica']  else 'No':<12} "
          f"{'Sí' if ev['es_normocromica'] else 'No':<13} "
          f"{'✓ SÍ' if ev['patron_completo'] else '✗ No'}")

# Filtrar solo los que cumplen el patrón
con_patron = [ev['paciente_id'] for ev in evaluaciones if ev['patron_completo']]
print(f"\nPacientes con anemia normocítica normocrómica: {con_patron if con_patron else 'Ninguno'}")
```
