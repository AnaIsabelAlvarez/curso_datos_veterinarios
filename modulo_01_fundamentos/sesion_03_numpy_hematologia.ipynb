## 1. Arrays: la estructura central de NumPy

Un array de NumPy es un bloque de memoria contigua con elementos del mismo tipo.
Es la diferencia entre una lista de cajas individuales y una tabla de valores empaquetados.

```python
# Crear un array desde una lista de Python — la forma más común
# np.array() es el constructor principal

hemoglobina = np.array([15.1, 11.8, 17.2, 9.5, 16.0, 12.5, 14.2, 10.2, 16.5, 13.8])

print(f"Array:      {hemoglobina}")
print(f"Tipo:       {type(hemoglobina)}")        # <class 'numpy.ndarray'>
print(f"dtype:      {hemoglobina.dtype}")        # float64 — tipo de cada elemento
print(f"Shape:      {hemoglobina.shape}")        # (10,) — 10 elementos, 1 dimensión
print(f"Ndim:       {hemoglobina.ndim}")         # 1 — número de dimensiones
print(f"Elementos:  {hemoglobina.size}")         # 10 — total de elementos
```

```python
# dtype — el tipo de dato del array
# NumPy elige el tipo automáticamente, pero puedes especificarlo

# Array de enteros
leucocitos_conteo = np.array([8500, 12300, 7200, 18600, 9100])
print(f"Leucocitos dtype: {leucocitos_conteo.dtype}")   # int64

# Array de floats (por defecto cuando hay decimales)
hematocrito = np.array([45.2, 35.4, 51.8, 28.6, 48.0])
print(f"Hematocrito dtype: {hematocrito.dtype}")        # float64

# Especificar dtype explícitamente — útil para optimizar memoria
# float32 usa la mitad de memoria que float64 (importante con millones de registros)
hematocrito_32 = np.array([45.2, 35.4, 51.8, 28.6, 48.0], dtype=np.float32)
print(f"float32 dtype: {hematocrito_32.dtype}")

# Array de strings — menos común en NumPy, pandas lo maneja mejor
razas = np.array(['Labrador', 'Beagle', 'Pastor Alemán'])
print(f"Strings dtype: {razas.dtype}")   # <U13 — Unicode string de hasta 13 chars
```

```python
# Funciones para crear arrays especiales — muy usadas en análisis

# np.zeros() — array de ceros, útil para inicializar estructuras
plantilla_resultados = np.zeros(10)
print(f"Zeros: {plantilla_resultados}")

# np.ones() — array de unos
factores_correccion = np.ones(5)
print(f"Ones:  {factores_correccion}")

# np.arange() — equivale a range() de Python pero devuelve array
# np.arange(inicio, fin, paso)
semanas = np.arange(1, 13, 1)      # semanas 1 a 12 de seguimiento
print(f"Semanas: {semanas}")

dias_seguimiento = np.arange(0, 30, 7)   # cada 7 días durante 30 días
print(f"Días de control: {dias_seguimiento}")

# np.linspace() — N puntos equidistantes entre inicio y fin
# Útil para curvas de calibración
concentraciones_calibrador = np.linspace(0, 100, 6)   # 6 puntos de 0 a 100 mg/dL
print(f"Calibradores: {concentraciones_calibrador}")
```

## 2. Operaciones vectorizadas — el poder real de NumPy

Con listas de Python: para sumar 5 a cada elemento, necesitas un bucle.
Con NumPy: `array + 5` lo hace en una sola instrucción, en C compilado.

```python
# Datos: hemoglobina de 10 pacientes caninos (g/dL)
hemoglobina = np.array([15.1, 11.8, 17.2, 9.5, 16.0, 12.5, 14.2, 10.2, 16.5, 13.8])

# Operaciones aritméticas — se aplican a CADA elemento del array
print("=== OPERACIONES VECTORIZADAS ===")
print(f"Original (g/dL):        {hemoglobina}")
print(f"En mmol/L (×0.621):     {np.round(hemoglobina * 0.621, 2)}")
print(f"Ajustado +1 g/dL:       {hemoglobina + 1.0}")
print(f"Al cuadrado:            {hemoglobina ** 2}")

# Comparación vectorizada — devuelve array de booleanos
# Esta es la base del filtrado en pandas
bajo_referencia  = hemoglobina < 12.0    # límite inferior canino
sobre_referencia = hemoglobina > 18.0    # límite superior canino

print(f"\nBajo referencia (<12):  {bajo_referencia}")
print(f"Sobre referencia (>18): {sobre_referencia}")

# Contar cuántos pacientes están fuera de rango
n_bajo = np.sum(bajo_referencia)    # sum() sobre booleanos: True=1, False=0
print(f"\nPacientes con Hgb baja: {n_bajo} de {len(hemoglobina)}")
```

```python
# Operaciones entre dos arrays — elemento a elemento
# Los arrays deben tener el mismo shape

eritrocitos = np.array([6.2, 4.8, 7.1, 3.9, 6.5, 5.2, 5.8, 4.1, 6.8, 5.5])
hemoglobina = np.array([15.1, 11.8, 17.2, 9.5, 16.0, 12.5, 14.2, 10.2, 16.5, 13.8])

# HCM = Hemoglobina (g/dL) / Eritrocitos (millones/µL) × 10
# Fórmula real usada en hematología para calcular HCM en pg
hcm_calculado = (hemoglobina / eritrocitos) * 10
print(f"HCM calculado (pg): {np.round(hcm_calculado, 1)}")

# Verificar rango de referencia para HCM canino (19.5–24.5 pg)
hcm_normal = (hcm_calculado >= 19.5) & (hcm_calculado <= 24.5)
print(f"HCM en rango normal: {hcm_normal}")
print(f"Todos normales: {np.all(hcm_normal)}")       # True solo si TODOS son True
print(f"Alguno normal:  {np.any(hcm_normal)}")       # True si AL MENOS UNO es True
```

```python
# Funciones matemáticas universales (ufuncs) de NumPy
# Se aplican elemento a elemento con la misma sintaxis

valores_glucosa = np.array([4.2, 8.9, 12.1, 3.8, 6.5, 15.3, 5.1, 9.7])

print(f"Glucosa (mmol/L):    {valores_glucosa}")
print(f"Log natural:         {np.round(np.log(valores_glucosa), 3)}")
print(f"Log base 10:         {np.round(np.log10(valores_glucosa), 3)}")
print(f"Raíz cuadrada:       {np.round(np.sqrt(valores_glucosa), 3)}")
print(f"Valor absoluto:      {np.abs(valores_glucosa - valores_glucosa.mean())}")

# np.where() — el if/else vectorizado
# np.where(condición, valor_si_True, valor_si_False)
# Equivale a un operador ternario aplicado a todo el array

estado_glucosa = np.where(
    valores_glucosa < 3.9,  'Hipoglucemia',
    np.where(
        valores_glucosa > 11.0, 'Hiperglucemia',
        'Normal'
    )
)

print(f"\nEstado glucémico: {estado_glucosa}")
```

## 3. Indexación — acceder a subconjuntos de datos

NumPy tiene tres formas de indexar: por posición, por rango (slicing) y por condición booleana.
La tercera es la más usada en análisis de datos y es exactamente lo que pandas usa internamente.

```python
hematocrito = np.array([45.2, 35.4, 51.8, 28.6, 48.0, 37.5, 42.6, 30.8, 49.5, 41.4])

# Indexación por posición — igual que listas de Python
print(f"Primer valor:    {hematocrito[0]}")
print(f"Último valor:    {hematocrito[-1]}")
print(f"Posiciones 2–4:  {hematocrito[2:5]}")    # slicing: [inicio:fin] fin no incluido
print(f"Cada dos:        {hematocrito[::2]}")     # [::paso]
print(f"Invertido:       {hematocrito[::-1]}")    # paso negativo = invertir

# Indexación por lista de posiciones
posiciones_interes = [0, 3, 7]
print(f"\nPacientes 0, 3, 7: {hematocrito[posiciones_interes]}")
```

```python
# Indexación booleana — la más importante para análisis de datos
# Misma mecánica que usaste en pandas en la Sesión 1

hematocrito = np.array([45.2, 35.4, 51.8, 28.6, 48.0, 37.5, 42.6, 30.8, 49.5, 41.4])

# Crear máscara booleana
mascara_anemia = hematocrito < 37.0
print(f"Máscara (anemia): {mascara_anemia}")

# Aplicar la máscara — devuelve SOLO los valores donde es True
hto_anemicos = hematocrito[mascara_anemia]
print(f"Hto de pacientes anémicos: {hto_anemicos}")

# Condición compuesta: anemia moderada (25–37%)
anemia_moderada = (hematocrito >= 25.0) & (hematocrito < 37.0)
print(f"Anemia moderada: {hematocrito[anemia_moderada]}")

# np.where() para obtener los ÍNDICES que cumplen la condición
indices_anemicos = np.where(mascara_anemia)[0]
print(f"Índices (posiciones) de pacientes anémicos: {indices_anemicos}")
```

## 4. Arrays 2D — matrices de resultados de laboratorio

Un array 2D es una tabla: filas = pacientes, columnas = analitos.
Esta es la estructura interna de un DataFrame de pandas.

```python
# Crear un array 2D: 5 pacientes × 4 analitos de la serie roja
# Cada fila es un paciente, cada columna es un analito

# Los datos como lista de listas (cada sublista = un paciente)
datos_serie_roja = np.array([
    # Erit,  Hgb,   Hto,   VCM
    [6.2,   15.1,  45.2,  72.9],   # C001
    [4.8,   11.8,  35.4,  73.8],   # C002
    [7.1,   17.2,  51.8,  72.9],   # C003
    [3.9,    9.5,  28.6,  73.3],   # C004
    [6.5,   16.0,  48.0,  73.8]    # C005
])

nombres_analitos  = ['Eritrocitos', 'Hemoglobina', 'Hematocrito', 'VCM']
nombres_pacientes = ['C001', 'C002', 'C003', 'C004', 'C005']

print(f"Shape: {datos_serie_roja.shape}")    # (5, 4) → 5 filas, 4 columnas
print(f"Ndim:  {datos_serie_roja.ndim}")     # 2 dimensiones
print(f"\nMatriz completa:\n{datos_serie_roja}")
```

```python
# Indexación en 2D: [fila, columna] — o [fila][columna]
# La coma dentro de [] es la forma NumPy — más eficiente

# Un paciente específico (fila completa)
print(f"Resultados C001 (fila 0): {datos_serie_roja[0, :]}")
print(f"Resultados C004 (fila 3): {datos_serie_roja[3, :]}")

# Un analito específico (columna completa)
print(f"\nTodas las hemoglobinas (col 1): {datos_serie_roja[:, 1]}")
print(f"Todos los hematocritos (col 2): {datos_serie_roja[:, 2]}")

# Un valor específico
hgb_c002 = datos_serie_roja[1, 1]
print(f"\nHemoglobina de C002: {hgb_c002} g/dL")

# Submatriz: primeros 3 pacientes, solo Hgb y Hto (columnas 1 y 2)
submatriz = datos_serie_roja[:3, 1:3]
print(f"\nHgb y Hto de C001–C003:\n{submatriz}")
```

```python
# Estadísticas sobre ejes — clave para análisis por analito o por paciente
# axis=0: operar sobre filas (resultado por columna/analito)
# axis=1: operar sobre columnas (resultado por fila/paciente)

print("=== ESTADÍSTICAS POR ANALITO (axis=0) ===")
medias_analitos = datos_serie_roja.mean(axis=0)
for nombre, media in zip(nombres_analitos, medias_analitos):
    print(f"  {nombre:<15}: {media:.2f}")

print("\n=== MÍNIMO POR ANALITO ===")
mins = datos_serie_roja.min(axis=0)
for nombre, minimo in zip(nombres_analitos, mins):
    print(f"  {nombre:<15}: {minimo:.2f}")

print("\n=== MEDIA POR PACIENTE (axis=1) ===")
# Menos útil clínicamente, pero ilustra el concepto
medias_pacientes = datos_serie_roja.mean(axis=1)
for paciente, media in zip(nombres_pacientes, medias_pacientes):
    print(f"  {paciente}: media de analitos = {media:.2f}")
```

## 5. Estadística con NumPy — control de calidad analítico

El control de calidad interno de un laboratorio se basa en estadística descriptiva 
aplicada a los resultados del material de control. NumPy tiene todas las herramientas 
necesarias para automatizar esta evaluación.

```python
# Simulación: resultados del material de control para hemoglobina
# 20 mediciones del mismo control a lo largo de un mes
# Valor asignado del control: 14.5 g/dL ± 0.5 (límites ±2SD del fabricante)

np.random.seed(42)    # fijar semilla para reproducibilidad
# np.random.normal(media, desv_std, n_muestras)
control_hgb = np.random.normal(loc=14.5, scale=0.3, size=20)
control_hgb = np.round(control_hgb, 2)

print(f"Resultados del control (g/dL):\n{control_hgb}")
```

```python
# Estadísticas de control de calidad
media    = control_hgb.mean()
mediana  = np.median(control_hgb)
std      = control_hgb.std()         # desviación estándar
varianza = control_hgb.var()
cv_pct   = (std / media) * 100
minimo   = control_hgb.min()
maximo   = control_hgb.max()
rango    = maximo - minimo

# Percentiles — útiles para establecer límites de control
p25 = np.percentile(control_hgb, 25)
p75 = np.percentile(control_hgb, 75)
iqr = p75 - p25   # rango intercuartílico

print("=== ESTADÍSTICAS DE CONTROL — HEMOGLOBINA ===")
print(f"Media:                {media:.3f} g/dL")
print(f"Mediana:              {mediana:.3f} g/dL")
print(f"Desv. estándar:       {std:.3f} g/dL")
print(f"CV%:                  {cv_pct:.2f}%")
print(f"Mínimo / Máximo:      {minimo:.2f} / {maximo:.2f}")
print(f"Rango:                {rango:.3f}")
print(f"P25 / P75:            {p25:.3f} / {p75:.3f}")
print(f"IQR:                  {iqr:.3f}")
print(f"\nCriterio CV% < 3%: {'APROBADO ✓' if cv_pct < 3.0 else 'REVISAR ⚠'}")
```

```python
# Reglas de Westgard simplificadas con NumPy
# Las reglas de Westgard son el estándar internacional de control de calidad en laboratorio

valor_asignado = 14.5    # valor "verdadero" del material de control
sd_control     = std     # usamos la SD calculada del propio control

# Límites de control
limite_advertencia_sup = valor_asignado + 2 * sd_control    # +2SD
limite_advertencia_inf = valor_asignado - 2 * sd_control    # -2SD
limite_rechazo_sup     = valor_asignado + 3 * sd_control    # +3SD (regla 1:3s)
limite_rechazo_inf     = valor_asignado - 3 * sd_control    # -3SD

print("=== LÍMITES DE CONTROL WESTGARD ===")
print(f"Valor asignado:   {valor_asignado:.3f} g/dL")
print(f"SD del control:   {sd_control:.3f} g/dL")
print(f"Advertencia:      [{limite_advertencia_inf:.3f}, {limite_advertencia_sup:.3f}]")
print(f"Rechazo (1:3s):   [{limite_rechazo_inf:.3f}, {limite_rechazo_sup:.3f}]")

# Clasificar cada medición
fuera_advertencia = (control_hgb < limite_advertencia_inf) | (control_hgb > limite_advertencia_sup)
fuera_rechazo     = (control_hgb < limite_rechazo_inf)     | (control_hgb > limite_rechazo_sup)

print(f"\nMediciones con advertencia (>2SD): {np.sum(fuera_advertencia)}")
print(f"Mediciones con rechazo (>3SD):     {np.sum(fuera_rechazo)}")

if np.any(fuera_rechazo):
    indices_rechazo = np.where(fuera_rechazo)[0]
    print(f"  → Valores rechazados en posiciones: {indices_rechazo}")
    print(f"  → Valores: {control_hgb[fuera_rechazo]}")
else:
    print("  → Sin rechazos: control dentro de límites ✓")
```

```python
# Visualización: gráfico de Levey-Jennings
# El gráfico estándar de control de calidad en laboratorio clínico

dias = np.arange(1, len(control_hgb) + 1)

fig, ax = plt.subplots(figsize=(12, 5))

# Línea de los valores del control
ax.plot(dias, control_hgb, 'o-', color='steelblue', 
        linewidth=1.5, markersize=6, label='Control Hgb', zorder=3)

# Línea de la media
ax.axhline(y=media, color='green', linewidth=1.5, 
           linestyle='-', label=f'Media ({media:.3f})')

# Líneas de advertencia ±2SD
ax.axhline(y=limite_advertencia_sup, color='orange', linewidth=1.2,
           linestyle='--', label=f'+2SD ({limite_advertencia_sup:.3f})')
ax.axhline(y=limite_advertencia_inf, color='orange', linewidth=1.2,
           linestyle='--', label=f'-2SD ({limite_advertencia_inf:.3f})')

# Líneas de rechazo ±3SD
ax.axhline(y=limite_rechazo_sup, color='red', linewidth=1.2,
           linestyle=':', label=f'+3SD ({limite_rechazo_sup:.3f})')
ax.axhline(y=limite_rechazo_inf, color='red', linewidth=1.2,
           linestyle=':', label=f'-3SD ({limite_rechazo_inf:.3f})')

# Marcar puntos fuera de advertencia
if np.any(fuera_advertencia):
    ax.scatter(dias[fuera_advertencia], control_hgb[fuera_advertencia],
               color='orange', s=100, zorder=4, label='Advertencia')

# Zona aceptable sombreada
ax.fill_between(dias, limite_advertencia_inf, limite_advertencia_sup,
                alpha=0.08, color='green', label='Zona ±2SD')

ax.set_xlabel('Día de control', fontsize=11)
ax.set_ylabel('Hemoglobina (g/dL)', fontsize=11)
ax.set_title('Gráfico de Levey-Jennings — Control de Calidad Hemoglobina\n'
             f'CV% = {cv_pct:.2f}% | n = {len(control_hgb)} mediciones', fontsize=12)
ax.legend(loc='upper right', fontsize=8, ncol=2)
ax.set_xlim(0.5, len(control_hgb) + 0.5)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('../datos/levey_jennings_hgb.png', dpi=150, bbox_inches='tight')
plt.show()
print("Gráfico guardado ✓")
```

## 6. NumPy y pandas — cómo se relacionan

pandas está construido sobre NumPy. Cada columna de un DataFrame es internamente 
un array de NumPy. Entender NumPy significa entender qué pasa por debajo de pandas.

```python
import pandas as pd

# Crear un DataFrame desde arrays de NumPy
eritrocitos  = np.array([6.2, 4.8, 7.1, 3.9, 6.5])
hemoglobina  = np.array([15.1, 11.8, 17.2, 9.5, 16.0])
hematocrito  = np.array([45.2, 35.4, 51.8, 28.6, 48.0])

df = pd.DataFrame({
    'eritrocitos_mill_uL': eritrocitos,
    'hemoglobina_g_dL':    hemoglobina,
    'hematocrito_pct':     hematocrito
})

# Extraer una columna de pandas como array de NumPy
hgb_array = df['hemoglobina_g_dL'].to_numpy()
print(f"Columna como array NumPy: {hgb_array}")
print(f"Tipo: {type(hgb_array)}")

# Los valores de una columna pandas son accesibles como array
print(f"\nValores internos de la columna: {df['hemoglobina_g_dL'].values}")

# Las operaciones NumPy funcionan directamente sobre columnas pandas
print(f"\nCV% de hemoglobina (vía NumPy sobre columna pandas):")
col_hgb = df['hemoglobina_g_dL']
cv = (col_hgb.std() / col_hgb.mean()) * 100
print(f"  {cv:.2f}%")
```

```python
leucocitos = np.array([8.5, 12.3, 7.2, 18.6, 9.1, 11.4, 8.8,
                        15.2, 7.9, 10.3, 4.8, 22.1, 6.5, 9.9, 13.7])

# 1. Estadísticas descriptivas
media   = leucocitos.mean()
mediana = np.median(leucocitos)
sd      = leucocitos.std()
cv      = (sd / media) * 100

print("=== ESTADÍSTICAS ===")
print(f"Media:   {media:.2f} miles/µL")
print(f"Mediana: {mediana:.2f} miles/µL")
print(f"SD:      {sd:.2f}")
print(f"CV%:     {cv:.1f}%")

# 2. Clasificación con np.where() anidado
clasificacion = np.where(
    leucocitos < 6.0,  'Leucopenia',
    np.where(
        leucocitos > 17.0, 'Leucocitosis',
        'Normal'
    )
)
print(f"\n=== CLASIFICACIÓN ===")
print(clasificacion)

# 3. Conteo por categoría
for categoria in ['Leucopenia', 'Normal', 'Leucocitosis']:
    n = np.sum(clasificacion == categoria)
    pct = (n / len(leucocitos)) * 100
    print(f"  {categoria:<14}: {n} pacientes ({pct:.0f}%)")

# 4. Pacientes fuera de rango
fuera_rango = (leucocitos < 6.0) | (leucocitos > 17.0)
indices_fuera = np.where(fuera_rango)[0]

print(f"\n=== FUERA DE RANGO ===")
print(f"Índices:  {indices_fuera}")
print(f"Valores:  {leucocitos[fuera_rango]} miles/µL")
print(f"Estados:  {clasificacion[fuera_rango]}")
```
